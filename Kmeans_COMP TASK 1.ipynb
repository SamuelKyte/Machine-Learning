# Apart from SepalLengthCm vs. SepalWidthCm, each scatter plot seems to be in three distinct clusters. Therefore I will select 'SepalLengthCm vs. PetalLengthCm' and 'PetalWidthCm vs. SepalWidthCm' because they have the most opposite cluster formations.

```python
# prepare data for training, note that we do not supply y in an unsupervised task (which is why it is called unsupervised).
X = df[['SepalLengthCm','PetalLengthCm']].values.reshape(-1,2)  

print('Data size:')
print(X.shape)
```

```python
def eval_Kmeans(x, k, r):
    kmeans = KMeans(n_clusters=k, random_state=r)
    kmeans.fit(x)    
    return kmeans.inertia_

def elbow_Kmeans(x, max_k=10, r=123):
    within_cluster_vars = [eval_Kmeans(x, k, r) for k in range(1,max_k+1)]
    plt.plot(range(1,11), within_cluster_vars,marker='o')
    plt.xlabel('K')
    plt.ylabel('Inertia')
    plt.show()

elbow_Kmeans(X)
```

```python
def scatter_Kmeans(x, k, r=123):
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=r)
    y_pred = kmeans.fit_predict(x)
    colours = 'rbgcmy'
    for c in range(k):
      plt.scatter(x[y_pred == c, 0], x[y_pred == c,1], c = colours[c], label = 'Cluster {}'.format(c))
      plt.scatter(kmeans.cluster_centers_[c, 0], kmeans.cluster_centers_[c, 1], marker='x', c = 'black')
    
    score = round(silhouette_score(X, kmeans.labels_, metric='euclidean'),2)
    plt.title('silhouette={}'.format(score), loc='right', fontdict={'fontsize': 16}, pad=-14)
    plt.xlabel('SepalLengthCm')
    plt.ylabel('PetalLengthCm')
    plt.legend()
    plt.show()

for k in range(4,7):
  scatter_Kmeans(X, k, r=0)
```
